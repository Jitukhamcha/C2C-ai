```python
#importing libraries
import pandas as pd
import sklearn
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report

import pydotplus

from IPython.display import Image
```

```python
dataset= pd.read_csv('golf.csv')
```

```python
dataset.head(3)
```

```python
dataset
```

```python
#one hot encoding
x=pd.get_dummies(dataset[['Outlook','Temperature','Humidity','Windy']])
```

```python
x
```

```python
y=dataset['Play']
```

```python
y
```

```python
#decision tree model
model = tree.DecisionTreeClassifier(
    criterion = 'entropy', random_state=100 # address ko lagi use garinxa randome_state
)
```

```python
model.fit(x,y)
```

```python
data_plot = tree.export_graphviz(model, out_file=None, feature_names = list(x.columns.values), class_names=['Not_play', 'Play'], 
                                 rounded=True, filled=True)
graph = pydotplus.graph_from_dot_data(data_plot)

Image(graph.create_png())
```
