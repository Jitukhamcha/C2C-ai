```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import tree
import pydotplus
from IPython.display import Image

from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
```

```python
#Reading data
dataset=pd.read_csv("C:\\Users\\Ghost\\Desktop\\AIML\\c2c-ai\\Assignment 3\\heart_failure_clinical_records_dataset.csv")
print(dataset.head())
```

```python
#splitting X and Y
x=dataset.iloc[:,:-1].values
y=dataset.loc[:,'DEATH_EVENT'].values
x_col=dataset.iloc[:,:-1]
y_col=dataset.loc[:,:'DEATH_EVENT']
```

```python
print("***** X-Values *****")
print(x)
print("***** It's Shape *****")
print(x.shape)
print("*****Y-Values*****")
print(y)
print("***** It's Shape *****")
print(y.shape)
```

```python
#Train_test Split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.2)
```

```python
#modeling our decision tre
model=tree.DecisionTreeClassifier(criterion="entropy",random_state=200)
```

```python
#fit our model
model.fit(x_train,y_train)
```

```python
data_plot = tree.export_graphviz(model, out_file=None, feature_names = list(x_col.columns.values), class_names=['Alive', 'Dead'], 
                                 rounded=True, filled=True)
graph = pydotplus.graph_from_dot_data(data_plot)

Image(graph.create_png())
```

```python
#Predicting on Test Datas
x_predict=model.predict(x_test)
print("x_predict vs y_test")
for i in range(0,len(x_predict)):
    print(f"{x_predict[i]} \t  {y_test[i]}")
#As train_test split accuracy decreases for test datas
```

```python

print(classification_report(y_test,x_predict))
```

```python

print(confusion_matrix(y_test,x_predict))
```

```python
print(accuracy_score(x_predict,y_test))
```
