```python
from keras_facenet import FaceNet
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
import cv2
import os
import numpy as np
import pickle
```

```python
dataset = "dataset/images"
```

## Initiazling facenet embeder

```python
embedder = FaceNet()
```

## Extracting Features

```python
features = []  
labels = []
for root, files, images in os.walk(dataset):
    print(root.split("/")[-1])

    for image in images:
        
        img = cv2.imread(os.path.join(root, image))
        embedeing = embedder.embeddings(np.expand_dims(img,axis=0))
        features.append(embedeing)
        labels.append(root.split("/")[-1])
```

```python
features = np.array(features)
fet = features[:,0,:]
labels = np.array(labels)
```

```python
fet.shape
```

```python
x_train, x_test , y_train, y_test =  train_test_split(fet, labels, test_size = 0.2, random_state=42)
```

```python
model = KNeighborsClassifier(n_neighbors=5)
model.fit(x_train, y_train)
```

```python
acc = model.score(x_test, y_test)
acc
```

```python
classes = ['badal',
 'bibek',
 'mukesh',
 'nasla',
 'prapti',
 'sujen',
 'swastika']
```

```python
for i in range(len(classes)):
    try:
        img = cv2.imread(f"dataset/images/{classes[i]}/{i}.png")
        emb=  embedder.embeddings(np.expand_dims(img,axis=0))
        result = model.predict(emb)
        result=result[0].split('\\')[1]
        print(f'real:{classes[i]}, prediction:{result}')
    except:
        img = cv2.imread(f"dataset/images/{classes[i]}/{i+1}.png")
        emb=  embedder.embeddings(np.expand_dims(img,axis=0))
        result = model.predict(emb)
        result=result[0].split('\\')[1]
        print(f'real:{classes[i]}, prediction:{result}')
```
